```python
import pylayers.gis.ezone as ez
import matplotlib.pyplot as plt
import numpy as np
```

`Ezone` class handles an earth zone which corresponds to the `strm` file. It has the same naming convention 
and corresponds to a portion of earth with 1° latitude and 1° of longitude. Let take an example with the file `N48W002.h5` 

```python
IlleVilaine = ez.Ezone()
IlleVilaine.load('N48W002.h5')
```

This file contains the srtm DEM data and a filtration of open street map database selecting only the building footprint. Let's have a look 

```python
f,a=IlleVilaine.show()
```

An earth zone as an attached dictionnary of "smart" cities, which contains the data of all the building of the city. Below is shown the example for Rennes. The building data is stored as masked array for fast manipulation and plotting. 

```python
IlleVilaine.dcity
```

## Zooming in 

For zooming into a smaller region, we define the zone to vizualize with 
`(lonmin,lonmax,latmin,latmax)`

```python
extent1 = (-1.7,-1.6,48.05,48.15)
extent1_cart  = ez.conv(extent1,IlleVilaine.m)
print "latlon extent :",extent1
print "Cartesian extent (meters):",extent1_cart
```

Once the selected extent has been chosen, it is possible to pass it to the `show` method for zooming in the map. 

```python
f,a = IlleVilaine.show(title='Rennes City Center',extent=extent1,bldg=True,height=False,contour=True)
```

The map shows above is graduated in longitude (horizontal axis) and latitude (vertical axis) but it is also possible to display it in cartesian coordinates as below

```python
f,a = IlleVilaine.show(title='Rennes City Center',extent=extent1_cart,coord='cartesian')
```

Let zoom to the University Rennes 1 campus.

```python
extent2 = (-1.655,-1.554,48.101,48.155)
extent2_cart = ez.conv(extent2,IlleVilaine.m)
f,a = IlleVilaine.show(title='Campus de Beaulieu',extent=extent2_cart,coord='cartesian')
```

## Profile extraction 

For predicting the radio propagation, it is important to retrieve the height profile between 2 points on the earth surface. The `profile` method does a profile extraction and geometrical calculation for further propagation loss calculation using the Deygout method. Points have to be expressed in (lon,lat) coordinates in WGS84 system.

```python
h,d,dh,nu,num,m,LOS = IlleVilaine.profile(pa=(-1.7,48.12),pb=(-1.2,48.2),fGHz=0.3)
```

```python
f = plt.figure(figsize=(15,5))
a=plt.plot(d,dh,'r',d,h,'b',d,m[0,:],'g',d,LOS,'k')
```

```python
f = plt.figure(figsize=(15,5))
a=plt.plot(d,nu)
a = plt.axis([0,40000,-2,2])
plt.title('Fresnel parameter')
plt.xlabel('Distance (meters)')
```
