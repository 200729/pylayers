```python
ip=EnImpulse(fe=100)
```

```python
fig = plt.figure(figsize=(10,10))
for k,snr in enumerate(range(30,-30,-10)):
    a = fig.add_subplot(3,2,k+1)
    ipn=ip.awgn(snr=snr,typ='snr')
    ipn.plot(typ='v',fig=fig,ax=a)
    a.set_title('SNR :'+str(snr)+' dB')
plt.tight_layout()
```
