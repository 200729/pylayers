```python
from pylayers.simul.link import * 
```

This notebook illustrates the `link` module. A `Dlink` is a deterministic SISO link

```python
L=DLink()
```

The `repr` function provides information about the default link, which is stored in the  `hdf5` file

```python
L
```

```python
aktk=L.eval()
```

The propagation channel (without antenna) can be vizualized on a ray by ray mode.

```python
C = L.C
C.show(cmap='jet')
```

It is possible to look at individual ray transfer function, as illustrated below.

```python
C.Ctt.y.shape
```

```python
plot(C.Ctt.x,abs(C.Ctt.y[0,:]))
```

```python
plot(C.Ctt.x,abs(C.Ctt.y[30,:]))
```

In the link we also have the transmission channel accounting for the effect of antennas and Friis factor

```python
plot(L.H.x,L.H.y[4,:]*4*pi*L.H.x/0.3)
```
