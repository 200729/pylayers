# PyLayers Overview




Ray Tracing tools can naturally bring the connection between radio channel observable and Tx/Rx positions.
The main objective is to study dedicated evolutions for Ray Tracing tools to address the positioning
simulation needs. It first consists in building the connection between the RT functional block and the
mobility model block. The second important evolution to introduce deals with
the capability to bring at once the radio channel observables on several MS with different air interfaces (IR-UWB and LTE)

Specific objectives : Bring to Ray Tracing radio positioning simulator :

+ Multi node/Multi RAT capabilities
+ Specific evolution for IR-UWB waveforms
+ Mobility models for multi node dynamics

**RT simulation in the tracking loop**

*Adaptive RT - How to take advantage of a set of rays at one position to make a fast evaluation of the
set of rays at the next position?*

+ Evolution for dealing with IR UWB waveforms

## Exploiting Graph data structure

```python
L=Layout('WHERE1.ini')
try:
    L.dumpr()
except:
    L.build()
    L.dumpw()
```

Building the graph of non overlapping cycles

A problem of geometric graph theory $(\mathcal{G},\mathbf{p})$

+ sketch of the algorithm 

    1. Determine a basis of cycles : **nx.algorithms.cycle_basis()**
        - Fast algorithm $\mathcal{O}(N_e)$ purely topological, no need of metric properties.
    2. Simplification of cycles based on inclusion detection:
        - detect cycles sharing common segments 
        - split overlapping cycles (recursive implementation)
        - inclusion relation is not reciprocal
        $$ \textrm{if   } \mathcal{C}_1 \subset \mathcal{C}_2 \Rightarrow  Area(\mathcal{C}_1) < Area(\mathcal{C}_2)$$

## Graph of interactions

+ Each segment separates 2 cycles
+ Each segment (excepted ABSORBENT(0), METAL(-1) and AIR(1)) has 4 interactions
    + (nseg,ncy1)  R in cycle 1
    + (nseg,ncy2)  R in cycle 2
    + (nseg,ncy1,ncy2) T from cycle 1 to cycle 2
    + (nseg,ncy2,ncy1) T from cycle 2 to cycle 1

## Mobility : Using mobile agents

At a large scale, the movement of the agent can be summarized by successive transitions between rooms.
All the possible transitions between rooms are represented with the graph $\mathcal{G}_w$.

+ The graph $\mathcal{G}_w$ has the same room nodes $\mathcal{v}_r$ of $\mathcal{G}_r$, plus some extra door nodes. 
+ Those extra nodes are added to encourage the agent both to exit by the door and to avoid unreal trajectory e.g. go to the center of a corridor instead of directly reaching the opposite room. 
+ Those extra door nodes are named as such the segment their belong.


Nodes are connected via transition segments such as doors.
