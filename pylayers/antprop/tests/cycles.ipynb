A signature is calculated between two cycles. This presents an interest for evaluating coverage maps. All links which have their termination in the same cycle can be obtained with the same signature. 

```python
nc1 = 7
nc2 = 33
```

The topological graph $\mathcal{G}_t$ contains all the cycles of the layout. A member data of the node of $\mathcal{G}_t$ is the shapely polygon associated with the cycle. 

```python
poly1 = L.Gt.node[nc1]['polyg']
cp1 = poly1.centroid.xy

poly2 = L.Gt.node[nc2]['polyg']
cp2 = poly2.centroid.xy
```

Here the centroïd of each cycle is chosen in order to become respectively the transmitter and the receiver of the link. 

```python
ptx = np.array([cp1[0][0],cp1[1][0],1.5])
prx = np.array([cp2[0][0],cp2[1][0],1.5])
print 'Tx : ',ptx
print 'Rx : ',prx
v   = prx-ptx
mv  = np.sqrt(sum(v*v,axis=0))
vn  = v/mv
print vn
```

```python
d = np.sqrt(np.dot((ptx-prx),(ptx-prx)))
tau = d/0.3
print "Distance (m): ,",d
print "Delay (ns): ,",tau
```

Il faut trouver les interactions de la ligne de front

```python
for cy in lcil:
    nghb = nx.neighbors(L.Gt,cy)
    print cy,nghb
    poly1 = L.Gt.node[cy]['polyg']
    cp1 = poly1.centroid.xy
    p1  = np.array([cp1[0][0],cp1[1][0],1.5])
    figure()
    for cya in nghb:
        poly2 = L.Gt.node[cya]['polyg']
        cp2 = poly2.centroid.xy
        p2 = np.array([cp2[0][0],cp2[1][0],1.5])
        vp = p2-p1
        m2 = sqrt(sum((p2-p1)*(p2-p1),axis=0))
        vpn = vp/m2
        dp = dot(vpn,vn)
        if dp>0:
            poly2.plot('g')
        else:
            poly2.plot('r')
        axis([-25,30,0,18])
```
