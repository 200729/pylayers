```python
from pylayers.antprop.antenna import *
from pylayers.antprop.antvsh import *
#from mayavi import mlab
```

```python
def err(A,Fth,Fph,kf,dsf,theta,phi):
    """

    """
    st    = outer(sin(theta),ones(len(phi)))    
    dTh   = (FTh[kf,:,:]-A.Ftheta[kf,::dsf,::dsf])
    dPh   = (FPh[kf,:,:]-A.Fphi[kf,::dsf,::dsf])
    dTh2  = real(dTh*conj(dTh))*st
    dPh2  = real(dPh*conj(dPh))*st
#
    vTh2  = real(A.Ftheta[kf,::dsf,::dsf]*conj(A.Ftheta[kf,::dsf,::dsf]))*st 
    vPh2  = real(A.Fphi[kf,::dsf,::dsf]*conj(A.Fphi[kf,::dsf,::dsf]))*st
#
    mvTh2 = sum(vTh2)
    mvPh2 = sum(vPh2)
#
    errTh = sum(dTh2) 
    errPh = sum(dPh2) 
#
    errelTh = errTh/mvTh2
    errelPh = errPh/mvPh2
    errel   = (errTh+errPh)/(mvTh2+mvPh2)
    
    return(errelTh,errelPh,errel)
```

#Electrical Delay Determination (Author Troels Pedersen)

```python
delayCandidates  =  arange(-10,10,0.001)
freq = A.fa.reshape(104,1,1)
maxPowerInd      = unravel_index(argmax(abs(A.Ftheta)),shape(A.Ftheta))
electricalDelay  = delayCandidates[argmax(abs(dot(A.Ftheta[:,maxPowerInd[1],maxPowerInd[2]],exp(2j*pi*freq.reshape(len(freq),1)*delayCandidates.reshape(1,len(delayCandidates))))))]

#plot(freq,angle(A.Ftheta[:,maxPowerInd[1],maxPowerInd[2]]*exp(2j*pi*freq.reshape(len(freq))*electricalDelay)))
disp('Electrical Delay = ' + str(electricalDelay)+' ns') 
```

# Electrical Delay compensation 

```python
# An example of broadcasting
A.Ftheta = A.Ftheta*exp(2*1j*pi*freq*electricalDelay)
A.Fphi   = A.Fphi*exp(2*1j*pi*freq*electricalDelay)
```

# Study of the quality of the reconstruction 

```python
# read file
dsf = 2


tn  = []
tet = []
tep = []
te  = []
tn  = range(1,10)
for n in tn:
    print 'n : ',n
    A.C.s1tos2(n)
    th      = A.theta[::dsf]
    ph      = A.phi[::dsf]
    Nt      = len(th)
    Np      = len(ph)
    Th      = kron(th,ones(Np))   
    Ph      = kron(ones(Nt),ph) 
    Fth,Fph = A.Fsynth2(Th,Ph)
    FTh = Fth.reshape(A.Nf,Nt,Np)
    FPh = Fph.reshape(A.Nf,Nt,Np)
    errelTh,errelPh,errel = err(A,Fth,Fph,20,dsf,th,ph)
    tet.append(errelTh)
    tep.append(errelPh)
    te.append(errel)
```

```python
plot(array(tn),10*log10(array(tep)),'b')
plot(array(tn),10*log10(array(tet)),'r')
plot(array(tn),10*log10(array(te)),'g')
xlabel('n')
ylabel('err_rel  (dB)')
#legend(('errel_phi','errel_theta','errel'))
```

```python
dsf = 2
k = 1
figsize(60,60)
A = Antenna('mat','S2R23.mat','ant/UWBAN/Matfile')
A.vshd(dsf)
Br2 = real(A.C.Br.s1*conj(A.C.Br.s1))
Ref = sum(Br2,axis=2)
terr   = []
for filename in serie2:
     # read file
     A = Antenna('mat',filename,'ant/UWBAN/Matfile')
     #print A.AntennaName   
     # apply vsha (downsampled by dsf) 
     A.vshd(dsf)
     Br2 = real(A.C.Br.s1*conj(A.C.Br.s1))
     ImBr2 = abs(sum(Br2,axis=2)-Ref)
     errtot = sum(ImBr2)   
     terr.append(errtot)
     subplot(23,1,k)   
     #contourf(arange(104),arange(20),ImBr2[:,0:20].T,vmin=0,vmax=0.5,)
     imshow(ImBr2[:,0:20].T,vmin=0,vmax=0.06,interpolation="nearest")
     #imshow(ImBr2[:,0:20].T,interpolation="nearest")
     ylabel('n')
     colorbar()   
     title(A.AntennaName+' '+filename)   
     k=k+1
xlabel('freq index')        
```
