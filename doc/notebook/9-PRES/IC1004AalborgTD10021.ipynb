<h1 class="title">What is PyLayers ? </h1>

* An **open source** platform for **education and research**
* Writen is **Python**
* **Evolutive** 
    + git versioned
    + documented from the source code
    + unitested (although surely not enough)

* Organized as a set of specialized modules
    - [gis,antprop,simul,signal,mobility,util,...]
   
* **IPython notebook** oriented  
    + high interactivity
    + highly data oriented
    + Hierarchical data storage (*HDF5*)

<h1 class="title">Applications</h1>

* Prediction of radio observable for **localization in Indoor environments** 

* Analysis of channel measurements

* Adressing various radio standard including **UWB** radio 

* Progapagation modeling and **human mobility**

<h2> The Signature `.Si` </h2>

```python
Lk.Si.keys()
print Lk.Si[4].shape
print Lk.Si[4][0,:]
print Lk.Si[4][1,:]
```

+ Signatures are rays potentialities
+ A set of Signature is calculated from one Layout cycle to an other Layout cycle 
+ A Signature consists of 2 lists
    + A list of interaction entities (point (<0) or segment (>0))
    + A list of type of interactions (1:D, 2:R, 3:T)
+ A **same signature** is used for calculation of rays from one cycle to an other cycle
